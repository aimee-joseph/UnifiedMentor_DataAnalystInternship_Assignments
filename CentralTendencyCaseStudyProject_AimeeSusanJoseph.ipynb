### Observations

##### There are 4898 rows and 12 columns, where each row contains the details of the types of acids present in white-wine and the quality.

```python
plt.figure(figsize = (11, 6))
sns.histplot(data = data, x = "fixed acidity", color = "orange", edgecolor = "linen", alpha = 0.5, bins = 5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("Count")
```

##### The histogram is normally distributed. The values between 6 - 8 are highly dominant.

```python
print(round(data["fixed acidity"].mean(), 2))
```

```python
print(data["fixed acidity"].median())
```

```python
plt.figure(figsize = (11, 6))
sns.histplot(data = data, x = "fixed acidity", color = "orange", edgecolor = "linen", alpha = 0.5, bins = 5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("Count")
plt.vlines(data["fixed acidity"].mean(), ymin = 0, ymax = 4000, colors = "blue", label = "Mean")
plt.vlines(data["fixed acidity"].median(), ymin = 0, ymax = 4000, colors = "red", label = "Mean")
plt.legend()
```

##### The mean and the median are clear representations of the data. They are very close to each other. We can choose either of the parameters, such as mean, as the central measure of tendency.

```python
plt.figure(figsize = (11, 6))
sns.histplot(data = data, x = "volatile acidity", color = "green", edgecolor = "linen", alpha = 0.5, bins = 5)
plt.title("Histogram of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Count")
plt.show()
```

##### The histogram is not well distributed, it is skewed a little towards the right. Since skewness is observed, we can check the distribution using the "distplot" function.

```python
plt.figure(figsize = (11, 6))
sns.displot(data["volatile acidity"], color = "blue", kde = True)
plt.title("Distplot of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Density")
plt.show()
```

##### The above plot shows normal distribution, described by the mean and standard deviation. The normal distribution is refereed

```python
print(data["volatile acidity"].skew())
```

##### The skewness value is greater than 1, meaning that it is positively skewed.

```python
print(data["volatile acidity"].mean())
```

```python
print(data["volatile acidity"].median())
```

```python
plt.figure(figsize = (11, 6))
sns.histplot(data = data, x = "volatile acidity", color = "green", edgecolor = "linen", alpha = 0.5, bins = 5)
plt.title("Histogram of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Density")
plt.vlines(data["volatile acidity"].mean(), ymin = 0, ymax = 4000, colors = "blue", label = "Mean")
plt.vlines(data["volatile acidity"].median(), ymin = 0, ymax = 4000, colors = "red", label = "Median")
plt.legend()
plt.show()
```

##### The mean and the median are close to each other and the difference between them is very little. Hence we can choose the mean as the measure of central tendency.
