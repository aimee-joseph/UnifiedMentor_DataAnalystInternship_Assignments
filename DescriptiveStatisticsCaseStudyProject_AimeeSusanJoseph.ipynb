##### There are 50 rows and 6 columns, where each row contains the employee details of each employee in the company.

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = data, x = "age", y = "salary", color = "blue", edgecolor = "linen", alpha = 0.5)
plt.title("Scatterplot of Employee Age vs Employee Salary")
plt.xlabel("Employee Age")
plt.ylabel("Employee Salary")
plt.show()
```

##### In general, there is a relationship between age and salary. As the age increases, the salary increases. However, there are some outliers present.

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = data, x = "healthy_eating", y = "active_lifestyle", color = "red", edgecolor = "linen", alpha = 0.5)
plt.title("Scatterplot of Healthy Eating vs Active Lifestyle")
plt.xlabel("Healthy Eating")
plt.ylabel("Active Lifestyle")
plt.show()
```

##### In general, there is a relationship between healthy eating and the active lifestyle of an employee. As healthy eating increases, active lifestyle also increases.

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = data, x = "healthy_eating", y = "salary", color = "k", edgecolor = "linen", alpha = 0.5)
plt.title("Scatterplot of Healthy Eating vs Salary")
plt.xlabel("Healthy Eating")
plt.ylabel("Salary")
plt.show()
```

##### In general, there is a relationship between healthy eating and the salary of an employee. As healthy eating increases, salary also increases.

```python
plt.figure(figsize = (10, 5))
sns.countplot(data = data, x = "groups", edgecolor = "linen", alpha = 0.7)
plt.title("Countplot of Groups")
plt.xlabel("Groups")
plt.ylabel("Count")
plt.show()
```

##### Most employees belong to either blood group A or O, with A having the most frequency.

```python
plt.figure(figsize = (10, 5))
sns.histplot(data = data, x = "salary", color = "orange", edgecolor = "linen", alpha = 0.7, bins = 10)
plt.title("Histogram of Salary")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

##### The salaries are not uniform, which is not necessarily a discrepancy. We can observe that there are employees with salaries on either extremities of the histogram. We can see that the count of the highest earning employees is very less.

##### Overall observations:

1. As the employees' age increases, there is an increase in their salaries.

2. As the employees' healthy eating habit increases, there is an increase in their salaries.

3. The employee salary values are more spread out.

4. An employee having their healthy eating rate greater than 8, can be considered as 8 itself.

5. An employee with salary less than 1000, can be considered as a lower salary.

##### Extracting data based on thresholds:

Considering employees satisfying these conditions:

1. Employees having their healthy eating rates greater than 8.

2. Employees with salary less than 1000.

3. Employees with healthy eating rate greater than 8 and salary less than 1000.

```python
sub1 = data[data["healthy_eating"] > 8]
print(sub1)
```

```python
sub2 = data[data["salary"] < 1000]
print(sub2)
```

```python
sub = data[(data["healthy_eating"] > 8) & (data["salary"] < 1000)]
print(sub)
```

##### The only employee seemingly facing a discrepancy in salary as compared to healthy eating is the employee with ID = 26, who has a salary of 700.

##### From the given data, we can use simple visualisations to get a sense of how data is distributed. We can conduct preliminary analysis by subsetting data using well thought out thresholds and conditions.
